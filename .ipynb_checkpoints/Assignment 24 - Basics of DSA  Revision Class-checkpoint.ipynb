```python
def roman_to_integer(s):
    # Create a dictionary to store the values of Roman symbols
    roman_values = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }

    # Initialize the result
    result = 0

    # Iterate over the characters in the Roman numeral string
    for i in range(len(s)):
        # Get the value of the current symbol
        current_value = roman_values[s[i]]

        # If the current symbol is smaller than the next symbol,
        # subtract its value from the result
        if i + 1 < len(s) and roman_values[s[i + 1]] > current_value:
            result -= current_value
        else:
            # Otherwise, add its value to the result
            result += current_value

    return result

# Test the function with the given example
numeral = "III"
print("Roman numeral:", numeral)
print("Integer value:", roman_to_integer(numeral))
```

```python
def longest_substring_length(s):
    # Create a set to store unique characters in the current window
    unique_chars = set()

    # Initialize variables for the window boundaries and the result
    left = 0
    right = 0
    max_length = 0

    # Move the right boundary of the window
    while right < len(s):
        # If the current character is not in the window, add it to the set
        if s[right] not in unique_chars:
            unique_chars.add(s[right])
            # Update the maximum length if necessary
            max_length = max(max_length, right - left + 1)
            right += 1
        else:
            # If the current character is already in the window, remove the leftmost character
            unique_chars.remove(s[left])
            left += 1

    return max_length

# Test the function with the given example
s = "abcabcbb"
print("Input:", s)
print("Length of the longest substring without repeating characters:", longest_substring_length(s))
```

```python
def majority_element(nums):
    count = 0
    candidate = None

    # Find the potential majority element
    for num in nums:
        if count == 0:
            candidate = num

        count += 1 if num == candidate else -1

    # Verify if the potential majority element is the actual majority element
    count = 0
    for num in nums:
        if num == candidate:
            count += 1

    if count > len(nums) // 2:
        return candidate
    else:
        raise ValueError("No majority element found")

# Test the function with the given example
nums = [3, 2, 3]
print("Input:", nums)
print("Majority element:", majority_element(nums))
```

```python
def group_anagrams(strs):
    # Create a dictionary to store the groups of anagrams
    anagram_groups = {}

    # Iterate over each string in the array
    for s in strs:
        # Sort the string to use it as a key in the dictionary
        sorted_s = ''.join(sorted(s))

        # Add the string to its corresponding group
        if sorted_s in anagram_groups:
            anagram_groups[sorted_s].append(s)
        else:
            anagram_groups[sorted_s] = [s]

    # Convert the dictionary values to a list of lists
    result = list(anagram_groups.values())

    return result

# Test the function with the given example
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
print("Input:", strs)
print("Anagram groups:", group_anagrams(strs))
```

```python
def nth_ugly_number(n):
    ugly_numbers = [1] * n
    indices = [0] * 3  # Indices for 2, 3, and 5

    for i in range(1, n):
        # Calculate the next ugly number by selecting the minimum among the candidates
        next_ugly = min(ugly_numbers[indices[0]] * 2, ugly_numbers[indices[1]] * 3, ugly_numbers[indices[2]] * 5)

        ugly_numbers[i] = next_ugly

        # Update the indices for the candidates
        if next_ugly == ugly_numbers[indices[0]] * 2:
            indices[0] += 1
        if next_ugly == ugly_numbers[indices[1]] * 3:
            indices[1] += 1
        if next_ugly == ugly_numbers[indices[2]] * 5:
            indices[2] += 1

    return ugly_numbers[-1]

# Test the function with the given example
n = 10
print("Input:", n)
print("Nth ugly number:", nth_ugly_number(n))
```

```python
from collections import Counter

def top_k_frequent_words(words, k):
    # Count the occurrences of each word
    word_counts = Counter(words)

    # Sort the words based on frequency and lexicographical order
    sorted_words = sorted(word_counts, key=lambda word: (-word_counts[word], word))

    # Return the top k most frequent words
    return sorted_words[:k]

# Test the function with the given example
words = ["i", "love", "leetcode", "i", "love", "coding"]
k = 2
print("Input words:", words)
print("k:", k)
print("k most frequent words:", top_k_frequent_words(words, k))
```

```python
def max_sliding_window(nums, k):
  """
  Returns the max sliding window of size k for the given array.

  Args:
    nums: The array of integers.
    k: The size of the sliding window.

  Returns:
    A list of the max elements in the sliding window.
  """

  max_window = []
  window = []
  for i in range(len(nums)):
    if len(window) < k:
      window.append(nums[i])
    else:
      window.pop(0)
      window.append(nums[i])
    max_window.append(max(window))
  return max_window
r=max_sliding_window(nums,k)
nums=[1, 3, -1, -3, 5, 3, 6, 7]
k=3
print(r)
```

```python
def find_closest_elements(arr, k, x):
    left = 0
    right = len(arr) - 1

    # Binary search to find the index of the closest element to x
    while left < right:
        mid = (left + right) // 2
        if arr[mid] < x:
            left = mid + 1
        else:
            right = mid

    # Set the left and right pointers to form the window of k closest elements
    left -= 1
    right += 1

    # Find the k closest elements using the two-pointer approach
    while right - left - 1 < k:
        if left == -1:
            right += 1
        elif right == len(arr):
            left -= 1
        elif abs(arr[left] - x) <= abs(arr[right] - x):
            left -= 1
        else:
            right += 1

    # Return the k closest elements
    return arr[left + 1:right]

# Test the function with the given example
arr = [1, 2, 3, 4, 5]
k = 4
x = 3
print("Input arr:", arr)
print("k:", k)
print("x:", x)
print("k closest elements:", find_closest_elements(arr, k, x))
```
