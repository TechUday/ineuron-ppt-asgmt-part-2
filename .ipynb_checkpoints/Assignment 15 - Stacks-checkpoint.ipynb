```python
def nextGreaterElements(arr):
    n = len(arr)
    result = [-1] * n
    stack = []

    for i in range(n - 1, -1, -1):
        while stack and stack[-1] <= arr[i]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(arr[i])

    return result


arr = [1, 3, 2, 4]
result = nextGreaterElements(arr)
print(result)
```

```python
arr = [6, 8, 0, 1, 3]
result = nextGreaterElements(arr)
print(result)
```

```python
def nearestSmallerElements(a):
    n = len(a)
    result = [-1] * n
    stack = []

    for i in range(n):
        while stack and stack[-1] >= a[i]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(a[i])

    return result


a = [1, 6, 2]
result = nearestSmallerElements(a)
print(result)
```

```python
a = [1, 5, 0, 3, 4, 5]
result = nearestSmallerElements(a)
print(result)
```

```python
class StackUsingQueues:
    def __init__(self):
        self.q1 = []
        self.q2 = []

    def push(self, x):
        self.q1.append(x)

    def pop(self):
        if not self.q1:
            return None
        
        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0))
        
        return self.q1.pop(0)

    def top(self):
        if not self.q1:
            return None
        
        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0))
        
        top_element = self.q1[0]
        self.q2.append(self.q1.pop(0))

        self.q1, self.q2 = self.q2, self.q1
        
        return top_element


stack = StackUsingQueues()
stack.push(2)
stack.push(3)
print(stack.pop())  
stack.push(4)
print(stack.pop())  
```

```python
def reverseStack(stack):
    if not stack:
        return

    top = stack.pop()
    reverseStack(stack)
    insertAtBottom(stack, top)


def insertAtBottom(stack, item):
    if not stack:
        stack.append(item)
        return

    top = stack.pop()
    insertAtBottom(stack, item)
    stack.append(top)

St = [3, 2, 1, 7, 6]
reverseStack(St)
print(St) 
```

```python
def reverseString(s):
    stack = []
    reversed_string = ""

    # Push each character onto the stack
    for char in s:
        stack.append(char)

    # Pop each character from the stack and append it to the reversed string
    while stack:
        reversed_string += stack.pop()

    return reversed_string



S = "Hello, World!"
reversed_S = reverseString(S)
print(reversed_S)  
```

```python
S = "Geek for Geeks!!"
reversed_S = reverseString(S)
print(reversed_S)  
```

```python
def evaluatePostfixExpression(S):
    stack = []

    for char in S:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()

            result = performOperation(operand1, operand2, char)
            stack.append(result)

    return stack.pop()


def performOperation(operand1, operand2, operator):
    if operator == '+':
        return operand1 + operand2
    elif operator == '-':
        return operand1 - operand2
    elif operator == '*':
        return operand1 * operand2
    elif operator == '/':
        return operand1 / operand2



S = "231*+9-"
result = evaluatePostfixExpression(S)
print(result)  
```

```python
S = "123+*8-"
result = evaluatePostfixExpression(S)
print(result)  
```

```python
class MinStack:
    def __init__(self):
        self.stk1 = []
        self.stk2 = [float('inf')]

    def push(self, val: int) -> None:
        self.stk1.append(val)
        self.stk2.append(min(val, self.stk2[-1]))

    def pop(self) -> None:
        self.stk1.pop()
        self.stk2.pop()

    def top(self) -> int:
        return self.stk1[-1]

    def getMin(self) -> int:
        return self.stk2[-1]

min_stack = MinStack()

min_stack.push(-2)
min_stack.push(0)
min_stack.push(-3)


min_value = min_stack.getMin()
print(min_value)  # -3

min_stack.pop()

top_value = min_stack.top()
print(top_value)  # 0

min_value = min_stack.getMin()
print(min_value)  # -2
```

```python
def trap(height):
  """
  Given an elevation map, compute the amount of water that can be trapped after raining.

  Args:
    height: A list of non-negative integers representing the elevation of the map.

  Returns:
    The amount of water that can be trapped.
  """

  n = len(height)
  left_max = [0] * n
  right_max = [0] * n

  # Find the maximum height from the left for each index.
  for i in range(1, n):
    left_max[i] = max(left_max[i - 1], height[i])

  # Find the maximum height from the right for each index.
  for i in range(n - 2, -1, -1):
    right_max[i] = max(right_max[i + 1], height[i])

  # Calculate the amount of water that can be trapped at each index.
  water = 0
  for i in range(n):
    water += min(left_max[i], right_max[i]) - height[i]

  return water
height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]

water = trap(height)

print(water)
```
