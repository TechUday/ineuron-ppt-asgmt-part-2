```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def find_max_subtree_sum(root):
    max_sum = float('-inf')

    def dfs(node):
        nonlocal max_sum
        if node is None:
            return 0

        left_sum = dfs(node.left)
        right_sum = dfs(node.right)

        # Calculate the sum of the subtree rooted at the current node
        subtree_sum = left_sum + right_sum + node.val

        # Update the maximum sum if the current subtree has a higher sum
        max_sum = max(max_sum, subtree_sum)

        # Return the sum of the subtree rooted at the current node
        return subtree_sum

    dfs(root)

    return max_sum
# Create a binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Find the maximum sum of a subtree
max_sum = find_max_subtree_sum(root)
print("Maximum sum of a subtree:", max_sum)
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def construct_bst(level_order):
    if not level_order:
        return None

    root = TreeNode(level_order[0])
    queue = [root]
    i = 1

    while i < len(level_order):
        current_node = queue.pop(0)
        left_val = level_order[i] if i < len(level_order) else None
        right_val = level_order[i + 1] if i + 1 < len(level_order) else None

        if left_val is not None:
            current_node.left = TreeNode(left_val)
            queue.append(current_node.left)

        if right_val is not None:
            current_node.right = TreeNode(right_val)
            queue.append(current_node.right)

        i += 2

    return root
# Create a level order traversal list
level_order = [4, 2, 5, 1, 3]

# Construct the BST
root = construct_bst(level_order)

# Verify the BST
# In-order traversal should give a sorted list
def inorder_traversal(node):
    if node is None:
        return []
    return inorder_traversal(node.left) + [node.val] + inorder_traversal(node.right)

result = inorder_traversal(root)
print("Constructed BST:", result)
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def check_level_order_bst(arr):
    if not arr:
        return True

    stack = []
    root = TreeNode(float('inf'))
    stack.append(root)

    for i in range(1, len(arr)):
        node = TreeNode(arr[i])

        if arr[i] < stack[-1].val:
            stack[-1].left = node
        else:
            while stack and arr[i] > stack[-1].val:
                last_popped = stack.pop()

            if last_popped.val < arr[i]:
                return False

            last_popped.right = node

        stack.append(node)

    return True

# Given array
arr = [11,6,13,5,12,10]

# Check if the array represents a BST
is_bst = check_level_order_bst(arr)
print("Can represent a BST:", is_bst)
```
