```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def mergeTwoLists(l1, l2):
    if not l1:
        return l2
    if not l2:
        return l1

    if l1.val <= l2.val:
        l1.next = mergeTwoLists(l1.next, l2)
        return l1
    else:
        l2.next = mergeTwoLists(l1, l2.next)
        return l2


def mergeKLists(lists):
    if not lists:
        return None

    while len(lists) > 1:
        merged_lists = []
        for i in range(0, len(lists), 2):
            if i + 1 < len(lists):
                merged = mergeTwoLists(lists[i], lists[i + 1])
                merged_lists.append(merged)
            else:
                merged_lists.append(lists[i])

        lists = merged_lists

    return lists[0]
# Helper function to convert list representation to linked list
def createLinkedList(lst):
    head = ListNode(lst[0])
    curr = head
    for i in range(1, len(lst)):
        curr.next = ListNode(lst[i])
        curr = curr.next
    return head

# Helper function to convert linked list to list representation
def convertToList(head):
    lst = []
    curr = head
    while curr:
        lst.append(curr.val)
        curr = curr.next
    return lst

# Input lists as list representations
lists = [[1, 4, 5], [1, 3, 4], [2, 6]]

# Convert input lists to linked lists
linked_lists = [createLinkedList(lst) for lst in lists]

# Merge the linked lists
merged_list = mergeKLists(linked_lists)

# Convert the merged list back to a list representation
result = convertToList(merged_list)
print(result)
```

```python
def mergeAndCount(left, right):
    i, j = 0, 0
    merged = []
    count = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
            count += len(left) - i

    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged, count


def mergeSort(arr):
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr) // 2
    left, left_count = mergeSort(arr[:mid])
    right, right_count = mergeSort(arr[mid:])
    merged, merge_count = mergeAndCount(left, right)

    return merged, left_count + right_count + merge_count


def countSmaller(nums):
    _, counts = mergeSort(nums)
    return counts
nums = [5, 2, 6, 1]
result = countSmaller(nums)
print(result)
```

```python
def merge(left, right):
    merged = []
    i, j = 0, 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged


def mergeSort(nums):
    if len(nums) <= 1:
        return nums

    mid = len(nums) // 2
    left = mergeSort(nums[:mid])
    right = mergeSort(nums[mid:])
    return merge(left, right)


def sortArray(nums):
    return mergeSort(nums)
nums = [5, 2, 3, 1]
result = sortArray(nums)
print(result)
```

```python
def pushZerosToEnd(nums):
    left = 0
    right = 0

    while right < len(nums):
        if nums[right] != 0:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
        right += 1

    return nums
arr1 = [1, 9, 8, 4, 0, 0, 2, 7, 0, 6, 0]
result1 = pushZerosToEnd(arr1)
print(result1)  

arr2 = [1, 2, 0, 4, 3, 0, 5, 0]
result2 = pushZerosToEnd(arr2)
print(result2)  

```

```python
def rearrangeNumbers(nums):
    pos = 0
    neg = 0

    while pos < len(nums) and neg < len(nums):
        if nums[pos] > 0:
            pos += 1
        elif nums[neg] < 0:
            neg += 1
        else:
            nums[pos], nums[neg] = nums[neg], nums[pos]
            pos += 1
            neg += 1

    return nums
arr = [1, -3, 5, 6, -2, -8, -9, 10, 15]
result = rearrangeNumbers(arr)
print(result)
```

```python
def mergeArrays(arr1, arr2):
    i = 0
    j = 0
    k = 0
    merged = []

    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1
        k += 1

    while i < len(arr1):
        merged.append(arr1[i])
        i += 1
        k += 1

    while j < len(arr2):
        merged.append(arr2[j])
        j += 1
        k += 1

    return merged
arr1 = [1, 3, 4, 5]
arr2 = [2, 4, 6, 8]
result = mergeArrays(arr1, arr2)
print(result)  
```

```python
def intersection(nums1, nums2):
    set1 = set(nums1)
    result = []

    for num in nums2:
        if num in set1 and num not in result:
            result.append(num)

    return result
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]
result = intersection(nums1, nums2)
print(result)  
```

```python
def intersection(nums1, nums2):
    freqMap = {}
    result = []

    for num in nums1:
        freqMap[num] = freqMap.get(num, 0) + 1

    for num in nums2:
        if num in freqMap and freqMap[num] > 0:
            result.append(num)
            freqMap[num] -= 1

    return result
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]
result = intersection(nums1, nums2)
print(result)  
```
