```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def convert_binary_tree_to_bst(root):
    # Step 1: Perform in-order traversal and store node values
    def inorder_traversal(node):
        if not node:
            return []
        return inorder_traversal(node.left) + [node.val] + inorder_traversal(node.right)

    node_values = inorder_traversal(root)

    # Step 2: Sort the node values in ascending order
    node_values.sort()

    # Step 3 and 4: Update node values with sorted values
    index = 0

    def inorder_update(node):
        nonlocal index
        if not node:
            return
        inorder_update(node.left)
        node.val = node_values[index]
        index += 1
        inorder_update(node.right)

    inorder_update(root)

    return root
# Create a binary tree
root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)
root.right.left = TreeNode(6)
root.right.right = TreeNode(9)

# Convert the binary tree into a binary search tree
converted_root = convert_binary_tree_to_bst(root)

# Verify the converted binary search tree
# In-order traversal should give a sorted list
def inorder_traversal(node):
    if not node:
        return []
    return inorder_traversal(node.left) + [node.val] + inorder_traversal(node.right)

result = inorder_traversal(converted_root)
print("Converted BST:", result)
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def find_distance(root, key1, key2):
    if root is None:
        return 0

    if root.val > key1 and root.val > key2:
        return find_distance(root.left, key1, key2)

    if root.val < key1 and root.val < key2:
        return find_distance(root.right, key1, key2)

    if (root.val >= key1 and root.val <= key2) or (root.val <= key1 and root.val >= key2):
        return (
            distance_from_node(root, key1)
            + distance_from_node(root, key2)
        )

def distance_from_node(node, key):
    if node.val == key:
        return 0

    if node.val > key:
        return 1 + distance_from_node(node.left, key)

    return 1 + distance_from_node(node.right, key)


# Create the BST
root = TreeNode(8)
root.left = TreeNode(3)
root.right = TreeNode(10)
root.left.left = TreeNode(1)
root.left.right = TreeNode(6)
root.right.right = TreeNode(14)
root.left.right.left = TreeNode(4)
root.left.right.right = TreeNode(7)
root.right.right.left = TreeNode(13)

# Find the distance between two nodes in the BST
key1 = 4
key2 = 7
distance = find_distance(root, key1, key2)
print("Distance between", key1, "and", key2, "is", distance)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def convert_binary_tree_to_doubly_linked_list(root):
    if root is None:
        return None

    current = root
    prev = None

    while current is not None:
        next = current.right
        current.right = prev
        prev = current
        current = next

    return prev

def main():
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)

    dll = convert_binary_tree_to_doubly_linked_list(root)

    while dll is not None:
        print(dll.data)
        dll = dll.right

if __name__ == "__main__":
    main()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.next = None

def connect_nodes_at_same_level(root):
    if root is None:
        return

    q = []
    q.append(root)
    prev = None

    while len(q) > 0:
        current = q.pop(0)

        if prev is not None:
            prev.next = current

        prev = current

        if current.left is not None:
            q.append(current.left)

        if current.right is not None:
            q.append(current.right)

def main():
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)

    connect_nodes_at_same_level(root)

    curr = root
    while curr is not None:
        print(curr.data, end=" → ")
        curr = curr.next

    print("-1")

if __name__ == "__main__":
    main()
```
