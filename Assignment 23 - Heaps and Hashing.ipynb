```python
def calculate_depth(preorder):
    if not preorder:
        return 0

    depth = 0
    for i in range(len(preorder)):
        if preorder[i] == "l":
            continue

        depth += 1
        depth = max(depth, calculate_depth(preorder[i + 1 :]) + 1)

    return depth

def main():
    preorder = "nlnnlllnlnl"

    print(calculate_depth(preorder))

if __name__ == "__main__":
    main()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def left_view(root):
    if root is None:
        return

    q = []
    q.append(root)
    res = []

    while len(q) > 0:
        current = q.pop(0)
        res.append(current.data)

        if current.left is not None:
            q.append(current.left)

    return res

def main():
    root = Node(4)
    root.left = Node(5)
    root.right = Node(2)
    root.left.left = Node(3)
    root.left.right = Node(6)
    root.right.left = Node(1)
    root.right.right = Node(7)

    print(left_view(root))

if __name__ == "__main__":
    main()
```

```python
# Definition of a binary tree node
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def right_view(root):
    if root is None:
        return []

    # Use a queue for level order traversal
    queue = [(root, 0)]
    right_view_nodes = []

    while queue:
        node, level = queue.pop(0)

        # Check if it's the rightmost node at the current level
        if level == len(right_view_nodes):
            right_view_nodes.append(node.data)

        # Add the right child to the queue first
        if node.right:
            queue.append((node.right, level + 1))

        # Add the left child to the queue
        if node.left:
            queue.append((node.left, level + 1))

    return right_view_nodes

# Create the binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.right.right = Node(8)

# Print the right view of the binary tree
print("Right view of the tree is:", ' '.join(map(str, right_view(root))))
```

```python
# Definition of a binary tree node
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.hd = 0  # Horizontal distance

def bottom_view(root):
    if root is None:
        return []

    # Use a queue for level order traversal
    queue = [(root, 0)]
    bottom_view_map = {}

    while queue:
        node, hd = queue.pop(0)

        # Update the horizontal distance of the current node
        node.hd = hd

        # Update the node in the bottom view map
        bottom_view_map[hd] = node.data

        # Add the left child to the queue with updated horizontal distance
        if node.left:
            queue.append((node.left, hd - 1))

        # Add the right child to the queue with updated horizontal distance
        if node.right:
            queue.append((node.right, hd + 1))

    # Extract the bottom view nodes in the correct order
    bottom_view_nodes = [bottom_view_map[hd] for hd in sorted(bottom_view_map.keys())]

    return bottom_view_nodes

# Create the binary tree
root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)

# Print the bottom view of the binary tree
print("Bottom view of the tree is:", ' '.join(map(str, bottom_view(root))))
```
