```python
def firstUniqChar(s):
    char_count = {}  # Hashmap to store character counts

    # Count occurrences of each character
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    # Find the first non-repeating character and return its index
    for i, char in enumerate(s):
        if char_count[char] == 1:
            return i

    return -1  # If no non-repeating character found


s = "leetcode"
print(firstUniqChar(s))  
```

```python
def max_subarray_sum(nums):
    max_sum = float('-inf')
    current_sum = 0

    for num in nums:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)

    return max_sum

def min_subarray_sum(nums):
    min_sum = float('inf')
    current_sum = 0

    for num in nums:
        current_sum = min(num, current_sum + num)
        min_sum = min(min_sum, current_sum)

    return min_sum

def maxSubarraySumCircular(nums):
    max_sum = max_subarray_sum(nums)
    total_sum = sum(nums)
    min_sum = min_subarray_sum([-num for num in nums])

    if total_sum == min_sum:
        return max_sum

    return max(max_sum, total_sum - min_sum)


nums = [1, -2, 3, -2]
print(maxSubarraySumCircular(nums))  
```

```python
from collections import deque

def countStudents(students, sandwiches):
    queue = deque(students)

    while sandwiches:
        if queue[0] == sandwiches[0]:
            queue.popleft()
            sandwiches.pop(0)
        else:
            if all(student == queue[0] for student in queue):
                break
            queue.append(queue.popleft())

    return len(queue)


students = [1, 1, 0, 0]
sandwiches = [0, 1, 0, 1]
print(countStudents(students, sandwiches))  
```

```python
from collections import deque

class RecentCounter:
    def __init__(self):
        self.requests = deque()

    def ping(self, t: int) -> int:
        self.requests.append(t)
        while self.requests[0] < t - 3000:
            self.requests.popleft()
        return len(self.requests)
recentCounter = RecentCounter()
print(recentCounter.ping(1))     
print(recentCounter.ping(100))   

```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def playGame(n, k):
    # Create circular linked list
    head = Node(1)
    current = head
    for i in range(2, n + 1):
        current.next = Node(i)
        current = current.next
    current.next = head

    # Simulate the game
    while current.next != current:
        for _ in range(k - 1):
            current = current.next
        current.next = current.next.next
        current = current.next

    return current.value


n = 5
k = 2
print(playGame(n, k))  
```

```python
from collections import deque

def deckRevealedIncreasing(deck):
    n = len(deck)
    deck.sort(reverse=True)  # Sort the deck in descending order

    result = [0] * n
    queue = deque(range(n))

    for card in deck:
        result[queue.popleft()] = card
        if queue:
            queue.append(queue.popleft())

    return result


deck = [17, 13, 11, 2, 3, 5, 7]
print(deckRevealedIncreasing(deck))
```

```python
class FrontMiddleBackQueue:
    def __init__(self):
        self.front = []
        self.back = []

    def pushFront(self, val):
        self.front.append(val)
        self.balance()

    def pushMiddle(self, val):
        middle = len(self.front) // 2
        self.front.insert(middle, val)
        self.balance()

    def pushBack(self, val):
        self.back.append(val)
        self.balance()

    def popFront(self):
        if self.front:
            return self.front.pop()
        elif self.back:
            return self.back.pop(0)
        else:
            return -1

    def popMiddle(self):
        if self.front:
            middle = (len(self.front) - 1) // 2
            return self.front.pop(middle)
        elif self.back:
            return self.back.pop(0)
        else:
            return -1

    def popBack(self):
        if self.back:
            return self.back.pop()
        elif self.front:
            return self.front.pop()
        else:
            return -1

    def balance(self):
        if len(self.front) > len(self.back) + 1:
            self.back.insert(0, self.front.pop())
        elif len(self.back) > len(self.front):
            self.front.append(self.back.pop(0))
q = FrontMiddleBackQueue()
q.pushFront(1)      
q.pushBack(2)       
q.pushMiddle(3)    
q.pushMiddle(4)     
print(q.popFront()) 
print(q.popMiddle())
print(q.popMiddle())
print(q.popBack())  
print(q.popFront()) 
```

```python
class DataStream:
    def __init__(self, value, k):
        self.value = value
        self.k = k
        self.stream = [None] * k
        self.index = 0
        self.count = 0

    def consec(self, num):
        self.stream[self.index] = num
        self.index = (self.index + 1) % self.k
        self.count = min(self.count + 1, self.k)

        if self.count < self.k:
            return False

        for i in range(self.k):
            if self.stream[i] != self.value:
                return False

        return True
dataStream = DataStream(4, 3)  
print(dataStream.consec(4))    
print(dataStream.consec(4)) 
print(dataStream.consec(4))    
print(dataStream.consec(3))    
```
