```python
def find_nearest_greater_frequency(arr):
    frequency = {}
    stack = []
    result = [-1] * len(arr)

    for i in range(len(arr)-1, -1, -1):
        frequency[arr[i]] = frequency.get(arr[i], 0) + 1

        while stack and frequency[arr[stack[-1]]] <= frequency[arr[i]]:
            index = stack.pop()
            result[index] = arr[i]

        stack.append(i)

    return result
arr = [1, 1, 2, 3, 4, 2, 1]
output = find_nearest_greater_frequency(arr)
print(output)
```

```python
def sort_stack(stack):
    temp_stack = []

    while stack:
        temp = stack.pop()

        while temp_stack and temp_stack[-1] > temp:
            stack.append(temp_stack.pop())

        temp_stack.append(temp)

    while temp_stack:
        stack.append(temp_stack.pop())

    return stack
stack = [34, 3, 31, 98, 92, 23]
output = sort_stack(stack)
print(output)
```

```python
def deleteMiddle(stack, k):
    if k == 1:
        stack.pop()
        return
    
    temp = stack.pop()
    deleteMiddle(stack, k - 1)
    stack.append(temp)

def deleteMiddleElement(stack):
    size = len(stack)
    middle = (size // 2) + 1
    deleteMiddle(stack, middle)

# Example usage:
stack = [1, 2, 3, 4, 5]
deleteMiddleElement(stack)
print(stack)  # Output: [1, 2, 4, 5]

stack = [1, 2, 3, 4, 5, 6]
deleteMiddleElement(stack)
print(stack)  # Output: [1, 2, 4, 5, 6]
```

```python
from queue import Queue

def check_queue_order(queue):
    stack = []
    second_queue = Queue()
    expected = 1

    while not queue.empty():
        if queue.queue[0] == expected:
            second_queue.put(queue.get())
            expected += 1
        elif stack and stack[-1] == expected:
            second_queue.put(stack.pop())
            expected += 1
        else:
            stack.append(queue.get())

    while stack and stack[-1] == expected:
        second_queue.put(stack.pop())
        expected += 1

    if queue.empty() and not stack:
        return "Yes"
    else:
        return "No"
queue = Queue()
queue.put(5)
queue.put(1)
queue.put(2)
queue.put(3)
queue.put(4)

output = check_queue_order(queue)
print(output)
```

```python
def reverse_number(number):
    number_str = str(number)
    stack = []

    for char in number_str:
        stack.append(char)

    reversed_str = ""

    while stack:
        reversed_str += stack.pop()

    reversed_number = int(reversed_str)

    return reversed_number
number = 12345
reversed_number = reverse_number(number)
print(reversed_number)
```

```python
from queue import Queue

def reverse_k_elements(queue, k):
    stack = []
    temp_queue = Queue()

    # Step 1: Dequeue the first k elements and push them into the stack
    for _ in range(k):
        stack.append(queue.get())

    # Step 2: Enqueue the remaining elements into the temporary queue
    while not queue.empty():
        temp_queue.put(queue.get())

    # Step 3: Pop elements from the stack and enqueue them back into the given queue
    while stack:
        queue.put(stack.pop())

    # Step 4: Enqueue the elements from the temporary queue back into the given queue
    while not temp_queue.empty():
        queue.put(temp_queue.get())

    return queue
queue = Queue()
queue.put(1)
queue.put(2)
queue.put(3)
queue.put(4)
queue.put(5)

k = 3
reversed_queue = reverse_k_elements(queue, k)

# Printing the elements of the reversed queue
while not reversed_queue.empty():
    print(reversed_queue.get())
```

```python
def count_remaining_words(sequence):
    stack = []

    for word in sequence:
        if stack and word == stack[-1]:
            stack.pop()
        else:
            stack.append(word)

    return len(stack)


sequence = ["ab", "aa", "aa", "bcd", "ab"]
result = count_remaining_words(sequence)
print(result)  


```

```python
def max_absolute_difference(arr):
    n = len(arr)
    left_smaller = [0] * n
    right_smaller = [0] * n
    stack = []

    # Finding the nearest smaller element on the right side
    for i in range(n):
        while stack and arr[stack[-1]] > arr[i]:
            right_smaller[stack.pop()] = i
        stack.append(i)

    stack.clear()

    # Finding the nearest smaller element on the left side
    for i in range(n - 1, -1, -1):
        while stack and arr[stack[-1]] > arr[i]:
            left_smaller[stack.pop()] = i
        stack.append(i)

    max_diff = 0

    # Computing the maximum absolute difference
    for i in range(n):
        diff = abs(left_smaller[i] - right_smaller[i])
        max_diff = max(max_diff, diff)

    return max_diff
arr = [2, 1, 8]
result = max_absolute_difference(arr)
print(result) 

```
